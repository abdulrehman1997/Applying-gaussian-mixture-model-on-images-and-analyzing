# Gaussian Mixture Model.

### In this Notebook we will apply Gaussian Mixture Model on images to segment them into different parts.

## Importing necessary packages.

```python
from PIL import Image
import numpy as np
import pandas as pd

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, calinski_harabasz_score

import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
%matplotlib inline
from mpl_toolkits import mplot3d

import plotly.express as px
```

## We have to import the Picture. 

### After importing the picture, the picture is preprocessed to be suitable to be input into the Gaussian Mixture Model.
### We have to do the following to the picture:
    - Convert to a NumPy Array.
    - Change to encoding to float32.
    - Normaalise the Array with MinMax Normalization.
    - Convert the 3 dimensional image array to a 5 dimensional feature vector. The 5-dimensions are normalized vertical and horizontal coordinates of the pixel relative to the top-left corner of the image and the normalized R, G and B values of the image at that pixel.

## Method to apply MinMax Normalisation to the NumPy array of the Image.

```python
def norm_pix(data):
    for i in range(data.shape[2]):
        mi = data[..., i].min() #finding the Minimum value data point in the data.
        ma = data[..., i].max() #finding the Maximum value data point in the data.
        # Now we Traverse over all the point and normalize them using the min max normalization formula
        # normalised = (x-min)/(max-min)
        for j in range(data.shape[0]):
            for k in range(data.shape[1]):
                data[j][k][i]=(data[j][k][i]-mi)/(ma-mi)
    return data
```

## Method to Convert the Image data to 5 features dataset with X,Y,R,G,B values as features.

- While converting to 5 features dataset, we take in the normalised R,G,B values. 
- The new X,Y values also have to be normalise for the new dataset to be fully normalised.
- In this function we Normalise the X,Y values before converting it into a 5 features dataset.

```python
def create_data(data): #Input is normalised Numpy array.
    shape = data.shape
    final = []
    for i in range(shape[0]):
        t1 = []
        for j in range(shape[1]):
            t2=[i/(shape[0]-1),j/(shape[1]-1)] # normalising the X,Y values and putting normslised values in list t2
            # Appending R,G,B values to t2 to get t2 = [X,Y,R,G,B]
            t2.append(data[i][j][0]) 
            t2.append(data[i][j][1])
            t2.append(data[i][j][2])
            t1.append(t2) # Appending t2 values of column j in row i into list t1 
        final.append(t1) # Appending t1 values which contain  t2 values of row i into list final
    # We now have a list of shape (data.shape[0],data.shape[1],5)
    # We need all [X,Y,R,G,B] values to be in one list.
    # We need list of shape (data.shape[0]*data.shape[1],5)
    data = []
    for i in range(len(final)):
        for j in range(len(final[0])):
            data.append(final[i][j])
    # Alternatively Reshape function can also be used.
    return data
```

## Method to Input an image and Preprocess the image to return a dataset with 5 features.

- Input the image
- Convert image to a NumPy Array.
- Change to encoding to float32.
- Normaalise the Array with MinMax Normalization.
- Convert the 3 dimensional image array to a 5 feature dataset.

```python
def preprocess(img):
    img = np.array(img) # Converting image img to a NumPy Array.
    img = img.astype('float32') # Changing the encoding of the numpy array img to float32.
    
    norm_img = norm_pix(img) # Normaalising the Array with MinMax Normalization.
    data = create_data(norm_img) # Converting the 3 dimensional norm_img to a 5 feature dataset.
    data = np.array(data) # Converting list to NumPy Array.
    
    return(data)
```

## Gaussian Mixture Model Analysis and image segmentation.

### Now that We have the image in a dataset format we can input this into the Gaussian Mixture Model and analyse the segments. 
### We have to do the following steps:
    - Choose a value for n (number of segments/components to cluster).
    - Input the dataset into the Gaussian Mixture Model and retreive the cluster labels for each datapoint.
    - Plot the datapoints with the cluster labels representing the color of the datapoints.
    - Run PCA on the dataset.
    - Plot the PCA with the cluster labels representing the color of the datapoints.
    - Run t-SNE on the dataset.
    - Plot the t-SNE with the cluster labels representing the color of the datapoints.
    - Calculate and save the n value, the silhouette coefficent and the Calinski Harabasz score for the clusters formed.

## Method to run the dataset through a Gaussian Mixture Model and plot the clusters retreived.

```python
def plot_gmm(data,n):
    
    gmm = GaussianMixture(n).fit_predict(data) # run the dataset through a Gaussian Mixture Model and retreive labels for n clusters.
    
    # Plotting the datapoints with the cluster label representing the color of the datapoint.
    plt.scatter(data[:,1],data[:,0],c=gmm) 
    plt.gca().invert_yaxis()
    plt.show()
    
    return gmm
```

## Method to run PCA on the dataset and return the Principal Components.

```python
def get_pca(data):
    
    # Applying PCA on Data to retreive the Principal Components.
    pca = PCA()
    principalComponents = pca.fit_transform(data)
    principalComponents = pd.DataFrame(data = principalComponents,
                                       columns = ['PC1','PC2','PC3','PC4','PC5'])
    
    return(pca,principalComponents)
```

## Method to plot the Principal Components with the labels retreived from the Gaussian Mixture Model.

While plotting the principal components we take the first two main principal components of the dataset.

```python
def plot_PCA_2D(principalComponents,labels):
    
    # Plotting the PCA with the cluster labels representing the color of the datapoint.
    plt.scatter(principalComponents['PC1'], principalComponents['PC2'],c=labels,s=1)
    plt.gca().invert_yaxis()
    plt.title('PCA 2D')
    plt.show()
```

## Method to run t-SNE on the dataset and return the 2 Entities.

```python
def get_tsne(data, n_components):
    
    # Applying t-SNE on Data to retreive the Entities.
    tsne = TSNE(n_components=n_components, perplexity=50)
    Entities = tsne.fit_transform(data)
    Entities = pd.DataFrame(data = Entities,columns = ['E1','E2'])
    return(tsne,Entities)
```

## Method to plot the Entities with the labels retreived from the Gaussian Mixture Model.

```python
def plot_tsne(Entities,labels):
    
    # Plotting the t-SNE with the cluster labels representing the color of the datapoint.
    plt.scatter(Entities['E1'], Entities['E2'],c=labels,s=1)
    plt.gca().invert_yaxis()
    plt.title('t-SNE Plot')
    plt.show()
```

Now that we have implemented all the basic functions required for the analysis we will now Implement a function to retrieve all the information.

## Method to analyse the clustering of a Gaussian Mixture Model on an image.

- Input the dataset into the Gaussian Mixture Model and retreive the cluster labels for each datapoint.
- Plot the datapoints with the cluster labels representing the color of the datapoints.
- Run PCA on the dataset.
- Plot the PCA with the cluster labels representing the color of the datapoints.
- Run t-SNE on the dataset.
- Plot the t-SNE with the cluster labels representing the color of the datapoints.
- Calculate and save the n value, the silhouette coefficent and the Calinski Harabasz score for the clusters formed.

```python
def gmm_analysis(data,n):
    
    gmm = plot_gmm(data,n) # Plotting the Gaussian Mixture Model clusters
    
    pca,principalComponents = get_pca(data) # Retreiving Principal Components
    plot_PCA_2D(principalComponents,gmm) # Plotting the PCA with Gaussian Mixture Model labels
    
    tsne , Entities = get_tsne(data,2) # Retreiving t-SNE Entities.
    plot_tsne(Entities,gmm) # Plotting the t-SNE with Gaussian Mixture Model labels
    
    sc = silhouette_score(data, gmm, metric='euclidean')
    ch = calinski_harabasz_score(data, gmm)
    
    # Calculating the silhouette coefficent.
    print('silhouette coefficent for n components = '+str(n)+' -> '+str(sc))
    
    # Calculating the Calinski Harabasz score.
    print('CH Index for n components = '+str(n)+' -> '+str(ch))
    
    return(sc,ch)
```

Now we can run the image through gmm_analysis() with n = 2 to try out this method.

```python
sc,ch = gmm_analysis(para,2)
```

We can see that The gaussian mixture model clusterd the parachute and the human as one cluster and the background as another cluster.
Now Lets try for n = 2,3,4,5,6,7 and analyse the results.

Let us save the silhouette coefficent and the Calinski Harabasz score for each value of n and plot the results to visualise the results.

```python
sc_ch = [] # Empty list to store silhouette coefficent and the Calinski Harabasz score for each value of n .
for n in [2,3,4,5,6,7]:
    
    print("For n = "+str(n))
    # Running gmm_analysis() on the image and retreiving the silhouette coefficent and the Calinski Harabasz score.
    sc,ch = gmm_analysis(para,n) 
    # appending the silhouette coefficent and the Calinski Harabasz score
    sc_ch.append([sc,ch,n])
```

## Method to plot the silhouette coefficent and the Calinski Harabasz results for various values of n.

```python
def plot_sc_ch(sc_ch):
    sc_ch = pd.DataFrame(data = sc_ch,columns = ['sc','ch','n'])
    fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(10,4))
    ax1.plot(sc_ch['n'], sc_ch['sc'])
    ax1.set_title('silhouette coefficent')
    ax2.plot(sc_ch['n'], sc_ch['ch'])
    ax2.set_title('CH Index')
    plt.show()
```

Now let us plot the silhouette coefficent and the Calinski Harabasz score for each value of n and visualise the results.

```python
plot_sc_ch(sc_ch)
```

After Comparing The silhouette coefficents and the Calinski Harabasz scores and looking at the plots we can see that n=5 is a good value as it has high coefficients and clusters very well. we can also see that the parachute has been totally isolated from the other clusers in the PCA.

n=2 is also a good value as it has a silhouette coefficent and clusters very well. But n = 5 gives the best silhouette coefficents and the Calinski Harabasz scores and clusters better.
